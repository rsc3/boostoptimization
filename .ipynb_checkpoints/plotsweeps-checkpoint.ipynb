```python
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
```

```python
flyingfox = pd.read_csv('data/front_flyingfox.csv')
platypus = pd.read_csv('data/front_platypus.csv')
redperson = pd.read_csv('data/front_redperson.csv')
```

```python
np.meshgrid(redperson.Mbps_UP)
```

```python
fig = plt.figure()
#ax = fig.gca(projection='3d')

# Make data.
X = redperson.Angle
Y = redperson.Azimuth
U = redperson.Mbps_UP

plt.tricontourf(X,Y,U)
plt.show()

D = redperson.Mbps_DOWN
plt.tricontourf(X,Y,D)
plt.show()

P = redperson.Ping
plt.tricontourf(X,Y,P)
plt.show()
```
